```python
import pandas as pd
import numpy as np
import requests, sys, webbrowser, bs4, serpapi
from serpapi.google_search_results import GoogleSearchResults
from bs4 import BeautifulSoup
import requests
import re


#Reading the CSV to load CSV into python for database manipulation.
labels = pd.read_csv('Candy Label Suppliers - Sheet1.csv')
labels
```

```python
#Removing the ".doc" froming the file name.
labels['Candy'] = labels['Candy'].str.replace(r'.doc', '')

#Concatenating the string.
labels["string"] = labels['Suppliers'] +" "+ labels['Candy'] + " ingredients"

#Declaring number per row so that we can iterate through the rows. 
rows = labels.iterrows()
iterator = 0
#Loop to search for web results
for rows in  labels.iterrows():
    
    iterator = iterator + 1
    
    try: 
        from googlesearch import search 
    except ImportError:  
        print("No module named 'google' found") 
    
    
    
    query = str(labels['string'][iterator])
  
    for j in search(query, tld="co.in", num=4, stop=4, pause=4): 
        
        try:
            response = requests.get(j)
            soup = BeautifulSoup(response.content, "html5lib")
            

            font = soup.find("b", text="Ingredients:")
            my_string=soup.get_text()

            my_string = my_string.split("Ingredients",1)[1]
            my_string = my_string.split(".",1)[0]
            n = "Results Found"
            
            print("\n" + str(iterator) +": "+ (str(labels['string'][iterator])))
            print(j)
            print(my_string[0:240])
            print(n)
                
        except:
            n = "No Results Found"
            print(str(iterator) +": "+ (str(labels['string'][iterator])))
            print(j)
            print(n)
        
    print("\n")
            
```
